```python
# define functions for reading and writing analog values and sending a TTL pulse using one of the digital channels

def lj_write(voltage,handle=lj_handle,channel=0):
    if channel not in [0,1]:
        raise Exception('Invalid LabJack AIN channel')
    name = 'DAC' + str(channel)
    value = voltage.to(u.volt).m
    ljm.eWriteName(lj_handle, name, value)
    
def lj_read(handle=lj_handle,channel=0):
    if channel not in [0,1,2,3]:
        raise Exception('Invalid LabJack AIN channel')
    name = 'AIN' + str(channel)
    return ljm.eReadName(lj_handle, name) * u.volt

def lj_ttl_pulse(handle=lj_handle,channel=0,wait_time=None):
    """Generate a TTL trigger pulse for other instruments waiting for a falling TTL edge."""
    if channel not in [0,1,2,3]:
        raise Exception('Invalid LabJack digital I/O channel')
    name = 'FIO' + str(channel)
    ljm.eWriteName(lj_handle, name, 1)
    if wait_time:
        sleep(wait_time.to(u.second).m)
    ljm.eWriteName(lj_handle, name, 0)
    

def lj_write_digital(state,handle=lj_handle,channel=0,wait_time=None):
    """Generate a TTL trigger pulse for other instruments waiting for a falling TTL edge."""
    if channel not in [0,1,2,3]:
        raise Exception('Invalid LabJack digital I/O channel')
    name = 'FIO' + str(channel)
    ljm.eWriteName(lj_handle, name, int(bool(state)))
```

## SRS SR570 Low-Noise Current Preamplifier

This is a listen only instrument controlled via serial (RS-232) so the visa address of the USB-serial adapter must be given explicitly to use the instrumental driver

```python
# open SR570 instrument
tia =  instrument(module='amplifiers.srs',classname='SR570',visa_address='ASRL6::INSTR')                
```

```python
# set up SR570
wait_time = 0.2
tia.set_filter_type('lp_12dB')
sleep(wait_time)
tia.set_sensitivity(100*u.pA/u.volt)
sleep(wait_time)
tia.set_lp_frequency(1 * u.Hz)
```

```python
# set up current measurements using LabJack Analog input 0 to read output of SR570
def sr570_current(wait_time=2*u.second,n_tries_max=10):
    current_resolved = False
    n_tries = 0
    while (not(current_resolved) and (n_tries<n_tries_max)):
        v = lj_read(channel=0)
        if abs(v) < (1*u.millivolt):
            old_sens = tia.sensitivity
            sleep(0.2)
            new_sens = tia.set_sensitivity(tia.sensitivity/3.0)
            print('low SR570 current signal, changing sensitivity from {:} to {:}'.format(old_sens, new_sens) )
            sleep(wait_time.to(u.second).m)
        elif ((abs(v) > (1*u.volt)) and (tia.sensitivity<(1*u.mA/u.volt))):
            old_sens = tia.sensitivity
            sleep(0.2)
            # do not set sensitivity to something higher than 1mA/volt
            new_sens = tia.set_sensitivity(np.min(np.array([tia.sensitivity.to(u.mA/u.volt).m*3.0,1])) * u.mA/u.volt)
            print('high SR570 current signal, changing sensitivity from {:} to {:}'.format(old_sens, new_sens) )
            sleep(0.2)
            tia.reset_filters()
            sleep(wait_time.to(u.second).m)
        else:
            current_resolved = True
            i = (tia.sensitivity * v).to(u.ampere)
            return i
        n_tries += 1
    return 0 * u.ampere
```

```python
# define trace acquisition functions

def collect_agilent_trace(ch=1,name='',data_dir='C:/Users/Lab/agilent_scope_data',return_data=False):
    timestamp_str = datetime.strftime(datetime.now(),'%Y_%m_%d_%H_%M_%S')
    fname = 'Agilent_54855A_scope_trace_' + name + '_' + timestamp_str + '.hdf5'
    fpath = path.join(data_dir,fname)
    SCOPE.take_traces(ch=ch)
    data = SCOPE.grab_traces(ch)[0]
    dt_s = data['dT']
    t0_s = data['T0']
    V = data['Y'] * u.volt  
    print('saving to '+fpath)
    with h5py.File(fpath, 'w') as f:
        #dset_t = f.create_dataset("t", (len(V),),dtype='float64',data=t.m,compression='gzip')
        dset_dt = f.create_dataset("dt", (1,),dtype='float64',data=dt_s)
        dset_t0 = f.create_dataset("t0", (1,),dtype='float64',data=t0_s)
        dset_V = f.create_dataset("V", (len(V),),dtype='float64',data=V.m,compression='gzip')
        f.flush()
    del data
    del dset_V
    SCOPE._write('RUN')
    if return_data:
        nt = len(V)
        t = np.linspace(t0_s,t0_s+nt*dt_s,len(V)) * u.second
        return t,V
    else:
        del V
        

def load_agilent_trace(name='',data_dir='C:/Users/Lab/agilent_scope_data',verbose=False):
    file_list =  glob(path.normpath(data_dir)+path.normpath('/Agilent_54855A_scope_trace_' + name + '_*'))
    latest_file = max(file_list,key=path.getctime)
    if verbose:
        print_statusline('Loading ' + name +' trace from file: ' + path.basename(path.normpath(latest_file)))
    with h5py.File(latest_file, 'r') as f:
        t0 = f['t0'][0] * u.second
        dt = f['dt'][0] * u.second
        V = f['V'][:] * u.volt
    nt = len(V)
    t = np.linspace(t0,t0+nt*dt,nt) * u.second
    return t,V


```

```python
# define functions to set the delayed time window for a "zoomed" acquisition

def collect_agilent_zoom(delay,ch=3,window_range=5*u.microsecond,name='',data_dir='C:/Users/Lab/agilent_scope_data',return_data=False):
    # record original settings
    pos0 = SCOPE._query(':TIMebase:POSition')
    range0 = SCOPE._query(':TIMebase:RANGe')
    sample_rate0 = SCOPE._query(':ACQuire:SRATe')
    # write zoom settings to scope
    SCOPE._write(':TIMebase:POSition {:1.3E}'.format(delay.to(u.second).m))
    SCOPE._write(':TIMebase:RANGe {:1.3E}'.format(window_range.to(u.second).m))
    SCOPE._write(':ACQuire:SRATe  MAX') # 20 billion samples per second... not bad
    SCOPE._write('RUN')
    sleep(0.5)
    # acquire data
    if return_data:
        t,V = collect_agilent_trace(ch=ch,name=name,data_dir=data_dir,return_data=True)
    else:
        collect_agilent_trace(ch=ch,name=name,data_dir=data_dir,return_data=False)
    # reset settings
    SCOPE._write(':TIMebase:POSition '+pos0)
    SCOPE._write(':TIMebase:RANGe '+range0)
    SCOPE._write(':ACQuire:SRATe '+sample_rate0)
    SCOPE._write('RUN')
    if return_data:
        return t,V
```

```python
# estimate normalization parameters in Ryan's paper so that I can compare experimental input power to calculated
# thresholds for free carrier and parametric oscillations

# see Appendix A in 'Conditions for Parametric and Free-Carrier Oscillation in Silicon Ring Cavities', Hamerly et al. JLT 2018

### device parameters ### (estiamted for now)

ring_diameter = 40 * u.um
ring_circumfrence = np.pi * ring_diameter
theta = 0.01
linewidth = 1 * u.GHz 
tau_ph = (1 / (2 * np.pi * linewidth)).to(u.ps)
tau_c = 30 * u.ps
A_eff = 0.1 * u.um**2

# using measured values from above:
# ng: 3.967
# fsr@1550nm: 601.3 GHz
# beta_2: 2.013 picosecond ** 2 / meter

FSR = 601.3 * u.GHz
finesse = (FSR / linewidth).to(u.dimensionless).magnitude
vg = (ring_circumfrence * FSR).to(u.meter / u.second)
ng = (u.speed_of_light / vg).to(u.dimensionless).magnitude
t_R = (ring_circumfrence * ng / u.speed_of_light).to(u.ps)
beta_2 = 2 * u.ps**2 / u.m

### material paramters ### (using values from Ryan's paper)
r = 0.189 # TPA/Kerr ratio
gamma = 3.1e-8 * u.cm / u.watt
mu = 25.0 # FCD / FCA ratio
sigma = 1.45e-17 * u.cm**2 # FCA cross section

### model normalization constants ###

xi_a = ( 2 * gamma * vg * tau_ph )**(-0.5)
xi_n = 1 / ( mu * sigma * vg * tau_ph )
xi_in = np.sqrt( t_R**2 / ( 8 * gamma * vg * tau_ph**3 * theta ) )
xi_t = np.sqrt( np.abs(beta_2) * vg * tau_ph )


### print calculated experimental parameters ###
print('Calculated experimental parameters:\n')
print('ng: {:1.2f}'.format(ng))
print('finesse: {:1.2f}'.format(finesse))

### print calculated model parameters ###
print('\nCalculated model parameters:\n')
print('tau_ph: {:3.2f} ns'.format(tau_ph.to(u.ns).m))
print('tau_c / tau_ph: {:3.2f}'.format((tau_c / tau_ph).to(u.dimensionless).m))
print('t_R: {:3.2f} ps'.format(t_R.to(u.ps).m))

print('xi_a^2: {:3.2f} W / cm^2'.format((np.abs(xi_a)**2).to(u.watt/u.cm**2).m))
print('xi_in^2: {:3.2f} W / cm^2'.format((np.abs(xi_in)**2).to(u.watt/u.cm**2).m))

print('xi_a^2*A_eff: {:3.2f} mW'.format((np.abs(xi_a)**2*A_eff).to(u.mW).m))
print('xi_in^2*A_eff: {:4.4f} uW'.format((np.abs(xi_in)**2*A_eff).to(u.uW).m))

print('xi_a / xi_in: {:3.2f}'.format(np.abs(xi_a / xi_in).to(u.dimensionless).m))
print('(xi_a / xi_in)^2: {:3.2f}'.format(np.abs(xi_a**2 / xi_in**2).to(u.dimensionless).m))


### calculate and print input power to reach S=1 ###
S_sq = 20
P_fiber_input_S_sq_equal_1 = ((np.abs(xi_in)**2*A_eff) / fiber_input_to_resonator_efficiency).to(u.mW)
P_fiber_input_threshold = P_fiber_input_S_sq_equal_1 * S_sq

print('\nThreshold power estimate:\n')


print('P_fiber_input_S_sq_equal_1: {:3.2f} mW'.format(P_fiber_input_S_sq_equal_1.to(u.mW).m))
print('S_sq: {:2.1f}'.format(S_sq))
print('P_fiber_input_threshold: {:3.2f} mW'.format(P_fiber_input_threshold.to(u.mW).m))
```

### EO Modulation sidebands

```python
# Added JGKB (now VersaWave) 40Gb/s polarization modulator between DFB laser's isolator and free space coupler.  
# part number EOM-PM-40G-5-1550-VP-0-FAP-FCA
# serial number 024REL02

# see test data report for this modulator in my Equipment and Software folder on labusers

modulator_throughput_efficiency = 2.12/13.44 # 0.158
print('modulator_throughput_efficiency: {:1.3f}'.format(modulator_throughput_efficiency))
modulator_throughput_efficiency_dB = 10*np.log10(modulator_throughput_efficiency) # -8.021 dB
print('modulator_throughput_efficiency_dB: {:1.3f}'.format(modulator_throughput_efficiency_dB))

# insertion loss at 1550nm is spec'd for this device at 6dB. I see 8dB, but I don't think they include
# butt-coupling efficiency in this spec. the extra two dB would mean 1dB loss per fiber couple (~80% efficiency each)
# which sounds plausible

## other measured specs for this modulator

# RF drive voltage for 40Gb/s "PRBS" modulation: 5.8V
# RF extinction ratio: 10.3 dBe
# DC extinction ratio at 10Hz: 16.2 dBe
# wavelength range: 1525-1565 nm

# I think this means I need ~5V  (P-P? probably) of microwave frequency power to get full polariztaion modulation
# Using 0A2CP12001-11 microwave VCO, I get ~dBm out at 8GHz. 
R = 50 * u.ohm
P_dBm = 12
P = 1 * u.mW * 10**(P_dBm/10.0)
V_mod_rms = np.sqrt(P * R).to(u.volt)
V_mod_pp = 2 * np.sqrt(2) * V_mod_rms
print('V_mod_rms: {:2.2f} V'.format(V_mod_rms.to(u.volt).m))
print('V_mod_pp: {:2.2f} V'.format(V_mod_pp.to(u.volt).m))
```

```python
# define data collection functions

def Vrb_sweep_fast_zooms(Vrb_list,delay_list,set_name,set_dir='C:/Users/Lab/agilent_scope_data'):
    for Vind, VV in enumerate(Vrb_list):
        if (VV.to(u.volt).m < 40.0):
            xhr_write(VV)
        else:
            raise Exception('attempted to write voltage above 40V to XHR')
        sleep(1)
        SCOPE._write('STOP')
        scope.stop_acquire()
        # collect agilent scope traces
        collect_agilent_trace(ch=1,name='Vrb{:2.1f}_agilent_ch1_P_trans_slow'.format(VV.m),data_dir=set_dir)
        collect_agilent_trace(ch=3,name='Vrb{:2.1f}_agilent_ch3_P_trans_fast'.format(VV.m),data_dir=set_dir)
        SCOPE._write('RUN')
        for delay in delay_list:
            collect_agilent_zoom(delay=delay,name='Vrb{:2.1f}_agilent_ch3_P_trans_fast_zoom_delay{:2.3f}'.format(VV.m,delay.to(u.ms).m),data_dir=set_dir)
        # collect Tek scope traces
        collect_scope_trace(channel=1,name='Vrb{:2.1f}_tek_ch1_P_trans_slow'.format(VV.m),data_dir=set_dir,return_data=False)
        collect_scope_trace(channel=2,name='Vrb{:2.1f}_tek_ch2_P_ref'.format(VV.m),data_dir=set_dir,return_data=False)
        collect_scope_trace(channel=3,name='Vrb{:2.1f}_tek_ch3_Vmod'.format(VV.m),data_dir=set_dir,return_data=False)
        collect_scope_trace(channel=4,name='Vrb{:2.1f}_tek_ch4_I_TPA'.format(VV.m),data_dir=set_dir,return_data=False)
        scope.run_acquire()
        
def Vrb_sweep(Vrb_list,set_name,set_dir='C:/Users/Lab/agilent_scope_data'):
    for Vind, VV in enumerate(Vrb_list):
        if (VV.to(u.volt).m < 40.0):
            xhr_write(VV)
        else:
            raise Exception('attempted to write voltage above 40V to XHR')
        sleep(1)
        scope.stop_acquire()
        # collect Tek scope traces
        collect_scope_trace(channel=1,name='Vrb{:2.1f}_tek_ch1_P_trans_slow'.format(VV.m),data_dir=set_dir,return_data=False)
        collect_scope_trace(channel=2,name='Vrb{:2.1f}_tek_ch2_P_ref'.format(VV.m),data_dir=set_dir,return_data=False)
        collect_scope_trace(channel=3,name='Vrb{:2.1f}_tek_ch3_Vmod'.format(VV.m),data_dir=set_dir,return_data=False)
        collect_scope_trace(channel=4,name='Vrb{:2.1f}_tek_ch4_I_TPA'.format(VV.m),data_dir=set_dir,return_data=False)
        scope.run_acquire()
```
